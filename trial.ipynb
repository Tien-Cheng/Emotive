```python
# filter emotions, page number, prediction per page, column sort, sort order

import itertools

emotion_list = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

fill_all = []
fil_base = []
fil_emotions = []

for pp in [3,9]:
    link_ext = '?'
    link_ext += f"per_page={pp}&"
    
    for cs in ["id", "emotion", "file_path"]:
        link_ext += f"col_sort={cs}&"
        for so in ["asc", "desc"]:
            link_ext += f"dir={so}&"
    
            fil_base.append(link_ext)

for i in range(1,8):
    for j in itertools.combinations(emotion_list, i):
        tmp = ''
        for e in j:
            tmp += f"c{e.capitalize()}=1&"
        fil_emotions.append(tmp)

for l in fil_base:
    for f in fil_emotions:
        fill_all.append((l+f)[:-1])

print(len(fill_all))
print(fill_all)
```
